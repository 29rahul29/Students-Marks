```python
import pandas as pd
df=pd.read_excel('student_data.xlsx',index_col='S.No')
df
```

```python
subjects=['Engineering Mathematics','Chemistry','Basic Electrical Engineering','Communication Skills','Basic Mechanical Engineering']
for subject in subjects:
    print(f"\n Top 3 students in {subject}: \n")
    top_students=df.sort_values(by=subject,ascending=False).head(3)
    print(top_students[['Name','Registration No',subject]])
```
